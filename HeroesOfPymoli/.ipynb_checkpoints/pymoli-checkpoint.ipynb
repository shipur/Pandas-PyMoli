```python
# Imports
import pandas as pd
import numpy as np
```

```python
# The path to the json file
file = "purchase_data.json"

# Read the purchase data into pandas
file_df = pd.read_json(file)
file_df.head()
```

```python
# Find the total number of players
totalPlayers = file_df["SN"].count()
print(totalPlayers)
#print("Total players: " + str(len(file_df["SN"].unique())))
```

```python
# Number of unique items
print("Number of unique items: " + str(len(file_df["Item Name"].unique())))
#Average purchase price
print("Average purchase price: " + str(file_df["Price"].mean()))
#Number of purchases
print("Number of purchases: " + str(file_df["Item ID"].count()))
#Total Revenue
print("Total Revenue: " + str(file_df["Price"].sum()))
```

```python
#Gender Demographics

print(file_df["Gender"].value_counts())
#Number of Males

# get indexes
file_df["Gender"].value_counts().index.tolist()

#Number of Males
maleCount = file_df["Gender"].value_counts().values.tolist()[0]
print("Number of males: " + str(maleCount))

#Percentage of Males
print("Percentage of Males: " + str(round((maleCount/totalPlayers * 100), 2))+"%")

#Number of Females
femaleCount = file_df["Gender"].value_counts().values.tolist()[1]
print("Number of females: " + str(femaleCount))

#Percentage of Females
print("Percentage of Females: " + str(round((femaleCount/totalPlayers * 100), 2))+"%")

#Number of Other/Non-Disclosed
otherCount = file_df["Gender"].value_counts().values.tolist()[2]
print("Number of Other/Non-Disclosed: " + str(otherCount))

#Percentage of Other/Non-Disclosed
print("Percentage of Other/Non-Disclosed: " + str(round((otherCount/totalPlayers * 100), 2))+"%")

print("---------------------------------------------------------------------------------")



#print("Males: " + str(male_df["Gender"].count()))
```

```python
#Age Demographics
# The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)

#First find the minimum and maximum age range
print(file_df["Age"].min())
print(file_df["Age"].max())

bins = [6,10,14,18,22,26,30,34,38,42,46]
group_labels = ["6-10 years","10-14 years", "14-18 years", "18-22 years", "22-26 years", "26-30 years",
               "30-34 years", "34-38 years", "38-42 years", "42-48 years"]

# Slice the data and place it into bins and place the data series into a new column inside of the DataFrame
file_df["Age Group"] = pd.cut(file_df["Age"],bins,labels=group_labels)
file_df.head(10)
```

```python
# Purchase Count
# Average Purchase Price
# Total Purchase Value
# Normalized Totals

# Create a GroupBy object based upon "Age Group"
file_group = file_df.groupby("Age Group")

# Find how many rows fall into each bin
print(file_group["Item ID"].count())

# Get the average of each column within the GroupBy object
file_group["Price"].mean()
file_group["Price"].sum()
Age_goup_df = pd.DataFrame({'Purchase Count' : file_group["Item ID"].count(), 
                        'Average Purchase Price': file_group["Price"].mean(),
                         'Total Purchase Value':file_group["Price"].sum()})
Age_goup_df.head()
```
